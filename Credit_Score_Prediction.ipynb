```python
from google.colab import drive
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import Model
from tensorflow.keras import Sequential
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import StackingRegressor
from sklearn.metrics import mean_squared_error
```

```python
#Importing the datasets
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/My Drive/sample_data/Projects/Credit_Score/credit_score.csv')
```

```python
df.head()
```

Doing data pre-processing

```python
df.drop(['CUST_ID'],inplace=True,axis=1)
```

```python
df.head()
```

```python
lb = LabelEncoder()
df['CAT_GAMBLING'] = lb.fit_transform(df['CAT_GAMBLING'])
```

```python
df.head()
```

```python
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(df[['CAT_GAMBLING']]).toarray())
# merge with main df bridge_df on key values
df = df.join(enc_df)
df.head()
```

```python
df.drop(['CAT_GAMBLING'],inplace=True,axis=1)
```

```python
df.rename(columns={0: "High_Gambling", 1: "Low_Gambling",2:"No_Gambling"},inplace=True)
```

```python
df.info()
```

```python
df = df[[col for col in df.columns if col != 'CREDIT_SCORE'] + ['CREDIT_SCORE']]
```

```python
df.head()
```

```python
x = df.values #returns a numpy array
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
```

```python
features = list(df.columns[:86])
Y = df[87]
X = df[features]

X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size = 0.7)
```

```python
individual_models = [SVR() , KNeighborsRegressor(), DecisionTreeRegressor()  ]
```

```python
for model in individual_models:
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  print("{} : RMSE : {:.3f}, Error Variance : {:.3f}".format(type(model).__name__ , np.sqrt(mean_squared_error(y_pred, y_test)), np.var(np.abs(y_test-y_pred))))
```

```python
base_models = [(type(i).__name__, i)for i in individual_models]
meta_model = LinearRegression()
```

```python
stacked_model = StackingRegressor(estimators = base_models , final_estimator = meta_model,
                                  cv = 3, n_jobs = -1, passthrough = False)

stacked_model.fit(X_train, y_train)
```

```python
y_pred = stacked_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
np.var(np.abs(y_test-y_pred))
```
